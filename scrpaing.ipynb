```python
import requests
from bs4 import BeautifulSoup

# URL to scrape
url = "https://timesofindia.indiatimes.com/world"

# Send a GET request to the webpage
response = requests.get(url)
response.raise_for_status()  # Raise an error if the request fails
# Parse HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")
print(soup)
# Find all article headlines (depends on website structure)
# For TOI, headlines often appear in <a> tags with specific classes
headlines = soup.find_all("a", class_="w_tle")  # You can change the class based on site

# Print scraped titles
print("Top Headlines:")
for headline in headlines:
    print("-", headline.get_text(strip=True))
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import os

download_dir="C:\g\cap_stone\kick_start\pp"
chrome_options = Options()
chrome_options.add_experimental_option("prefs", {
    "download.default_directory": download_dir,
    "plugins.always_open_pdf_externally": True, 
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
})
driver = webdriver.Chrome(options=chrome_options)
def downloading(doi):
    before_download = set(os.listdir(download_dir))
    driver.get("https://sci-hub.se/"+doi)  
    # for cookie in cookies:
    #     print(json.dumps(cookie))
    #     driver.add_cookie(json.dumps(cookie))

    driver.refresh()
    try:
        pdf_element = driver.find_element(By.ID, "pdf")

        pdf_url = pdf_element.get_attribute("src")

        driver.get(pdf_url)
        time.sleep(4) 

        after_download = set(os.listdir(download_dir))

        new_files = after_download - before_download
    except:
        new_files= ""
    return new_files
```

```python
import yfinance as yf

# Example: Get stock info for Apple
ticker = yf.Ticker("BLACKBUCK.NS")

# Get current price
current_price = ticker.info['currentPrice']
print("Current Price:", current_price)

# Get historical data
data = ticker.history(period="5d")
print(data)
```

```python
history = ticker.history(
    period="1d",          # Time period to download (alternative to start/end)
    interval="2m",         # Frequency of data
    start=None,            # Start date (YYYY-MM-DD)
    end=None,              # End date (YYYY-MM-DD)
    prepost=False,         # Include pre-market and after-hours data
    actions=True,          # Include dividends/splits
    auto_adjust=True,      # Adjust prices for splits/dividends
    repair=True,           # Fix missing data (recommended True)
    rounding=False,        # Round prices to 2 decimals
    timeout=10,            # Request timeout
    proxy=None,            # Proxy if behind a firewall
)
```
